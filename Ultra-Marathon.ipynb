```python
df["race_day"] = df["race_day"].astype(str).str.split(".").str.get(1).astype(int)
```

```python
df["race_day"] = (
    df["race_day"]
    .astype(str)
    .str.split(".")
    .str.get(1)
    .astype(float)  # Convert to float to handle NaN values
    .astype(int, errors='ignore')  # Convert to int, ignoring errors for NaN
)
```

```python
# Fill NaNs with a default value or drop rows
df["race_day"].fillna(0, inplace=True)  # Example of filling with 0
# Or
df.dropna(subset=["race_day"], inplace=True)  # Drop rows with NaN in 'race_day'
```

```python
df1.columns = df1.columns.str.strip()
```

```python
# Ensure race_day is numeric and within expected ranges
df1["race_day"] = pd.to_numeric(df1["race_day"], errors='coerce')

# Corrected season assignment logic
df1["race_season"] = df1["race_day"].apply(
    lambda x: "Winter" if x in [12, 1, 2] else 
              "Spring" if x in [3, 4, 5] else 
              "Summer" if x in [6, 7, 8] else 
              "Fall" if x in [9, 10, 11] else 
              "Unknown"  # Handle unexpected values
)
```
